### Step 3: Assignment (=) vs Comparison (==)
Remember: `=` stores a value, `==` asks a question.

```python
days = 28
print("Is days equal to 28?")
print(days == 28)   # This is a question -> True

print("Is days equal to 56?")
print(days == 56)   # This is a question -> False
```

### Step 4: Reassigning (Refilling the box)
A box can be refilled. The last value assigned always wins.

```python
status = "draft"
status = "final"
print(f"Current status: {status}")

# The String Gotcha
name = "  TERMINATION  "
name.strip()         # This creates a clean copy but doesn't save it!
print(f"After strip() without assignment: '{name}'")

name = name.strip()  # This puts the clean copy back in the box
print(f"After strip() WITH assignment: '{name}'")
```

### Why is this in Step 4?
In Python, strings are **immutable** (they cannot be changed in place).
When you run `name.strip()`, Python creates a brand new string, but the old `name` box still holds the messy text.

**The Rule:** If you want your variable to 'remember' the change, you **must** use the `=` operator to reassign that new value to the box name.

```python
deadline = "28 days"
deadline = "56 days"
print(f"The final deadline is: {deadline}")
```

```python
variable_box = "Initial Text"
variable_box = "Something totally different!"
print(f"The box now contains: {variable_box}")
```

### Step 5: Naming Conventions (Professional Habits)
In Python, we have rules for what you *can* name a variable and 'unwritten rules' (conventions) for what you *should* name them to look like a pro.

**The Rules:**
1. Must start with a letter or underscore `_` (Cannot start with a number).
2. Can only contain letters, numbers, and underscores (No spaces, no dashes `-`, no symbols like `!`).
3. Case-sensitive: `my_Var` and `my_var` are two different boxes.

**The Convention:**
Use `snake_case` (all lowercase with underscores between words) for readability.

### 💡 Key Concept: Assignment vs. Modification

In Step 4, we saw `deadline.strip()`. Here is the rule to remember:

1. **Operations (like `.strip()`)**: These just 'show' you a result. They do not change the variable itself.
2. **The Assignment Operator (`=`)**: This is the only way to actually change what is inside the 'box'.

**Why this matters:** If you clean a piece of data but don't reassign it (e.g., `text = text.strip()`), your program will keep using the old, messy data! Reassigning is how you 'save' your progress.

### Another Example: The 'Math' Gotcha
Just like `.strip()`, doing math on a variable doesn't change the variable unless you use `=`.

```python
price = 100

# 1. This calculates 120, but the result isn't stored anywhere!
price + 20
print(f"After math without assignment: {price}")

# 2. This takes the result (120) and puts it back into the price 'box'
price = price + 20
print(f"After math WITH assignment: {price}")
```

```python
price = 100
# Update the price immediately
price = price + 20

print(f"The final price is: {price}")
```

### Construction Example: The Change Order
In this scenario, we start with a base contract amount and use reassignment to update the total after a site variation.

```python
# 1. Initial Contract Value
contract_sum = 500000
print(f"Original Contract: ${contract_sum}")

# 2. A Change Order for additional foundation work arrives
change_order_01 = 15000

# 3. REASSIGNMENT: We update the box with the new total
# We take the old value, add the change, and save it back to 'contract_sum'
contract_sum = contract_sum + change_order_01

print(f"Revised Contract Sum after Change Order: ${contract_sum}")
```

**Why this matters for you:** If you just wrote `contract_sum + change_order_01` without the `contract_sum =` part, your 'box' would still say $500,000, and you would accidentally under-bill your client!

```python
contract_baseprice=500000
print(f"base price of contract is :${contract_baseprice}")
variation_order=100000
contract_total=contract_baseprice+variation_order
print(f"total contract price after variation:${contract_total}")
```
